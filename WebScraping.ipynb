```
from bs4 import BeautifulSoup
```

```
import requests
```

```
url= 'https://en.wikipedia.org/wiki/Coprinellus'
```

```
#we are sending the request to url and taking the response
response=requests.get(url)
response
```

```
response.text
```

```
#parsng library--->   html-parser   and lxml
Soup=BeautifulSoup(response.text, "lxml")
```

```
Soup
```

```
for Alink in Soup.find_all('a'):
  #print(Alink)
  print(Alink.get('href'))
```

```
for ptag in Soup.find_all('p'):
  #print(ptag)
  for btag in ptag.find_all("b"):
    print(btag.text)
```
